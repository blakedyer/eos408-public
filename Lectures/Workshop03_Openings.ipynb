### What are the essentially components of a good story?
<br>


<div class="split_panel" style='font-size:.8em'>

#### Fiction
- **Setup/Opening**: Who is the story about? Who are the characters? <span style='color:firebrick'>*Frodo, Gandalf, Sauron, Gollum..*</span>
- **Challenge/Conflict**: What needs to happen? <span style='color:firebrick'>*This is Sauron's one ring and it must be destroyed..*</span>
- **Action**: What are the steps to overcoming the challenge? <span style='color:firebrick'>*Leave The Shire, cross under the Misty Mountains, climb the stairs of Cirith Ungol.. etc*</span>
- **Resolution**: Without change, stories feel static. Someone or something is not the same at the end as at the beginning.. <span style='color:firebrick'>*The world is no longer in danger, characters have grown wiser..*</span>
    
</div>

<div class="split_panel" style='font-size:.8em;visibility:hidden'>
