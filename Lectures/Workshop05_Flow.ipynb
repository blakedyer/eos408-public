<h2>Why Flow Matters</h2>
<br>
  <p style="margin:.25em 0 .7em; font-size:1.05em;">
    Good flow means the reader always knows <strong>why this sentence or paragraph is here</strong>. They are motivated to keep reading.
  </p>

  <ul>
    <li>Flow lives at two levels:
      <ul>
        <li><strong>Paragraphs</strong>: one main idea that is clearly stated at the start.</li>
        <li><strong>Sentences</strong>: ideas progress from old → new, simple → complex.</li>
      </ul>
    </li>
    <li>Your task is to make the reader's task <strong>as easy as possible.</strong>.</li>
    <li>To do this, you need <strong>clear structure</strong> for the reader to follow.</li>
    <li style='color:firebrick'>The guidelines here are <b>not</b> hard-fast rules, and breaking them <i>intentionally</i> and <i>sparingly</i> can create the some of the <b>best</b> writing. However, ignoring these guidelines is not a good recipe... </li>
  </ul>

  <div style="margin-top:1em; background:#f6e58d; border:1px solid #f6e58d; border-radius:10px; padding:10px;">
    <strong>Key idea:</strong> If a reader only skimmed your first sentences in each paragraph, 
    they should still follow the story.
  </div>

<h2>Paragraphs: Argument First, Details Second</h2>
<br>
  <p style="margin:.25em 0 .7em;">
    A paragraph is a <strong>unit of argument</strong>, not just a block of text.
    The <strong>first sentence</strong> should state the main point.
  </p>

  <div style="display:grid; grid-template-columns:1fr 1fr; gap:14px;">
    <div style="border:1px solid #e5e7eb; border-radius:10px; padding:10px;">
      <h4 style="margin:.2em 0;">Role of the first sentence</h4>
      <ul style="margin:.4em 0;">
        <li>States the <strong>main idea</strong> of the paragraph.</li>
      </ul>
    </div>
    <div style="border:1px solid #e5e7eb; border-radius:10px; padding:10px;">
      <h4 style="margin:.2em 0;">What follows</h4>
      <ul style="margin:.4em 0;">
        <li>Evidence and examples.</li>
        <li>Logical reasoning.</li>
        <li>Limitations, caveats, and discussion.</li>
        <li>Sometimes a mini-wrap-up that links back to the bigger story or helps bridge to the next paragraph or section.</li>
      </ul>
    </div>
  </div>

<h2>Linking Paragraphs for Flow</h2><br>
    <div style="border:1px solid #f6e58d; background:#f6e58d; border-radius:10px; padding:10px;">
      <h4 style="margin:.2em 0;">A good diagnostic</h4>
      <ol style="margin:.4em 0; font-size:.95em;">
        <li>Read all the topic sentences in your peers work, ignoring everything else.</li>
        <li>Ask yourself: do they tell a clear, logical story?</li>
      </ol>
      <p style="margin:.4em 0; font-size:.9em;">
        If not, your paragraphs may be internally fine but globally out of order. Alternatively, your paragraphs many need to be internally reworked, divided, or combined.
      </p>
    </div>


 <h2>Sentence structure: given → new</h2>
 <br>
  <p style="margin:.25em 0 .7em;">
    Within a sentence, put <strong>given / familiar information first</strong> and 
    <strong>new / important information second</strong>.
  </p>

  <ul>
    <li><strong>Given / old:</strong> what the reader already knows from previous sentences or common context.</li>
    <li><strong>New:</strong> the fresh result, technical term, or key twist.</li>
    <li>This pattern:
      <ul>
        <li>Creates logical continuity.</li>
        <li>Puts emphasis on the <strong>end</strong> of the sentence.</li>
      </ul>
    </li>
  </ul>

  <div style="margin-top:1em; background:#f6e58d; border:1px solid #f6e58d; border-radius:10px; padding:10px;">
    <strong>Example:</strong><br>
    <em style='color:dodgerblue'>Given (old / familiar)</em> → <em style='color:firebrick'>New (new / important / technical)</em><br>
    “<span style='color:dodgerblue'>These <strong>observed sea-level differences</strong> are hard to explain</span>, and to account for them, <span style='color:firebrick'> we need a <strong>fully 3-D viscoelastic Earth model with laterally varying viscosity</strong>.</span>”
  </div>

<h2>Sentence order in practice</h2>
<br>
    <div style="border:1px solid #eb4d4b; background:#eb4d4b; border-radius:10px; padding:10px;">
      <h4 style="margin:.2em 0;">Example — Awkward order</h4>
      <p style="margin:.4em 0;">
        “A complex, fully 3-D viscoelastic Earth model with laterally varying viscosity 
        is required to explain the observed sea-level differences among interglacial reef sites.”
      </p>
      <p style="margin:.4em 0; font-size:.9em;">
        Problem: heavy, technical phrase appears before we’ve reminded the reader what problem we’re solving.
      </p>
    </div>
  <div style="margin-top:1em; background:#6ab04c; border:1px solid #6ab04c; border-radius:10px; padding:10px;">
    <h4 style="margin:.2em 0;">Example — A solution</h4>
    <p style="margin:.3em 0;"><strong>After:</strong> “To explain the observed sea-level differences, we need a <strong>complex, fully 3-D viscoelastic Earth model with laterally varying viscosity</strong>.”</p>
    <p style="margin:.3em 0; font-size:.9em;">
      Same content, but the familiar concept (“observed sea-level differences”) comes earlier,
      and the heavy technical phrase moves to the end position for emphasis and better flow.
    </p>
  </div>
